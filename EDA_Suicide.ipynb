**IMPORTING LIBRARIES AND READING THE DATA**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/eda data science 2.csv")
```

```python
df
```

**Processing the data for EDA**

```python
df.isnull().sum() #checking for null values
```

```python
df.drop_duplicates()  #dropping the duplicates
```

```python
df.State.unique() #AllIndia, States and Uts can be removed from the data set as it can lead to misimnformation 
```

```python
df = df.drop(df[(df.State == 'Total (Uts)') & (df.State == 'Total (All India)') &
               (df.State == 'Total (States)')].index)
df.State.unique()
```

```python
#df.Year.nunique()
df.Year.unique()
```

```python
df.Type_code.unique()
```

```python
df.Type.unique()
```

```python
df=df.drop(df[(df.Type =='By Other means')|(df.Type=='Other Causes (Please Specity)')| (df.Type=='Others (Please Specify)')|(df.Type=='Causes Not known')].index)
df = df.drop(df[df['Total'] == 0].index)
```

```python
df.Gender.unique()
```

```python
plt.figure(figsize=(15,10))
df_st = df.groupby('State').sum().Total.sort_values(ascending=False)
sns.barplot(y=df_st.index,x=df_st.values,orient='h')
plt.ylabel('State')
plt.title('Suicide stat State-wise')
```

Maharashtra is the state with most suicides followed by West Bengal and TamilNadu 

```python
df.Age_group.value_counts()
#0-100+ has a lot of data so cannot be omited so, we create a seperate data frame which considers the age 
```

```python
df_age = df[df['Age_group']!='0-100+']
df_age
```

```python
df_age.Age_group.value_counts().plot(kind='barh' ,  color='k');
plt.title("Suicide Vs Age group")
plt.show()
```

Maximum Suicides are in the age group from 15 to 44

```python
df.Gender.value_counts().plot(kind='pie' , cmap= 'jet');
plt.title("Male Vs Female")
plt.show()
```

Suides in Males are more compared to the female gender 

```python
plt.figure(figsize=(14,8))
df_yr = df.groupby('Year').sum().Total
sns.lineplot(x=df_yr.index,y=df_yr.values,marker='o',markerfacecolor='r')
plt.xlabel('Year')
plt.title('Suicides each Year')
```

```python
df_ss=df[df['Type_code']=='Social_Status']
df_ss_tp = df_ss.groupby('Type').sum().Total
df_ss_tp = df_ss_tp.sort_values(ascending=False)
plt.figure(figsize=(14,8))
sns.barplot(x=df_ss_tp.index,y=df_ss_tp.values)
plt.xlabel('Type of Social Status')
plt.title('Suicide vs Social Status')
```

```python
df_es=df[df['Type_code']=='Education_Status']
df_es_tp = df_es.groupby('Type').sum().Total
df_es_tp = df_es_tp.sort_values(ascending=True)
plt.figure(figsize=(12,6))
sns.barplot(y=df_es_tp.index,x=df_es_tp.values,orient='h')
plt.xlabel('Education Status')
plt.title('Suicide vs Education Status')
```
